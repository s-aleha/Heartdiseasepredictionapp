```python
#task no 1
import numpy as np
import tensorflow as tf
from tensorflow import keras

X = np.array([[0],[1],[2],[3],[4],[5]], dtype=float)
y = np.array([[0],[0],[0],[1],[1],[1]], dtype=float)

model = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(6, activation='sigmoid'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(X, y, epochs=100, verbose=1)

prediction = model.predict(np.array([[3]], dtype=float))
print("Prediction for 3:", prediction)
```

```python
#task no 2
import numpy as np
from tensorflow import keras

X = np.array([[0],[1],[2],[3],[4],[5]], dtype=float)
y = np.array([[0],[0],[0],[1],[1],[1]], dtype=float)

model_tanh = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(6, activation='tanh'),
    keras.layers.Dense(1, activation='sigmoid')
])

model_tanh.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history_tanh = model_tanh.fit(X, y, epochs=100, verbose=1)
loss_tanh, accuracy_tanh = model_tanh.evaluate(X, y, verbose=0)
print("Tanh Model Accuracy:", accuracy_tanh)
```

```python
#task no 3
import numpy as np
from tensorflow import keras

X = np.array([[0],[1],[2],[3],[4],[5]], dtype=float)
y = np.array([[0],[0],[0],[1],[1],[1]], dtype=float)

model_relu_10 = keras.Sequential([
    keras.Input(shape=(1,)),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model_relu_10.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history_relu_10 = model_relu_10.fit(X, y, epochs=100, verbose=1)

loss_relu_10, accuracy_relu_10 = model_relu_10.evaluate(X, y, verbose=0)
print("ReLU (10 neurons) Accuracy:", accuracy_relu_10)
```
