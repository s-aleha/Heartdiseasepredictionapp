```python
#task no 1
import numpy as np
from sklearn.linear_model import Perceptron

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,1])

model = Perceptron()

model.fit(X, y)

prediction = model.predict([[1,1]])

print("Prediction:", prediction)
```

```python
#task no 2
import numpy as np
from sklearn.linear_model import Perceptron

X = np.array([[1],
              [2],
              [4],
              [5]])

y = np.array([0, 0, 1, 1])

model = Perceptron()

model.fit(X, y)

prediction = model.predict([[3]])

print("Prediction for 3 hours:", prediction)
```
